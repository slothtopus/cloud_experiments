```python
import math
import matplotlib.pyplot as plt

precision = 1000
x_vals = []
y_vals = []
thetas = []
for i in range(0,precision):
    i = i/precision
    theta = i*2*math.pi
    thetas.append(theta)
    x_vals.append(math.cos(theta))
    y_vals.append(math.sin(theta))

def fract(x):
    return x - math.floor(x)


def map_xy(x,y):
    theta = math.acos(x)
    v = fract(math.acos(x)/(0.5*math.pi))
    if y > 0:
        return v
    else:
        return 1 - v


def cubic_pulse(c, w, x):
    x = abs(x-c)
    if x > w:
        return 0
    x = x / w
    return 1 - x*x*(3-2*x)
    
plt.plot(thetas, [cubic_pulse(0.5, 0.1, map_xy(x,y)) for (x,y) in zip(x_vals, y_vals)])




"""
p = 100
x_vals = [i / p for i in range(p)]
y_vals = [cubic_pulse(0.5, 0.1, x) for x in x_vals]


plt.plot(x_vals, y_vals)
"""
```
